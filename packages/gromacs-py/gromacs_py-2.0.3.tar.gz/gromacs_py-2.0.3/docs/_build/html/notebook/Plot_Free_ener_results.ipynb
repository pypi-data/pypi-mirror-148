```python
import re

data_str ="""
Y  |  p-Cresol       | c1ccccc1O      |0|-6.13 ± 0.60| NaN ± NaN|-8.78  | -7.59 ± 0.23|  1.94 | NaN ± NaN
F  |  toluene        | c1ccccc1       |0|-0.90 ± 0.20| NaN ± NaN|-4.49  | -3.32 ± 0.11|  2.72 | NaN ± NaN
W  |  3-methylindole | c1cccc2c1Ncc2C |0|-5.88 ± NaN | NaN ± NaN|-9.45  |-10.02 ± 0.21|  2.60 | NaN ± NaN
A  |  methane        | C              |0| 2.00 ± 0.20| NaN ± NaN| 0.49  |  0.91 ± 0.05|  1.06 | NaN ± NaN
V  |  propane        | CCC            |0| 2.00 ± 0.20| NaN ± NaN|-1.25  | -0.87 ± 0.07|  2.36 | NaN ± NaN
I  |  butane         | CCCC           |0| 2.15 ± NaN | NaN ± NaN|-1.87  | -1.66 ± 0.09|  2.93 | NaN ± NaN
L  |  isobutane      | CC(C)C         |0| 2.30 ± 0.20| NaN ± NaN|-1.46  | -1.48 ± 0.07|  2.72 | NaN ± NaN
M  |  methyl ethyl sulfide | CCSC     |0|-1.50 ± 0.20| NaN ± NaN|-3.50  | -2.94 ± 0.08|  1.47 | NaN ± NaN
S  |  methanol       | CO             |0|-5.10 ± 0.60| NaN ± NaN|-4.58  | -3.24 ± 0.22| -0.35 | NaN ± NaN
T  |  ethanol        | CCO            |0|-5.00 ± 0.60| NaN ± NaN|-4.06  | -3.43 ± 0.15| -0.60 | NaN ± NaN
N  |  acetamide      | CC(=O)N        |0|-9.71 ± 0.60| NaN ± NaN|-7.99  | -8.44 ± 0.18| -1.25 | NaN ± NaN
Q  |  propionamide   | CCC(=O)N       |0|-9.40 ± 0.20| NaN ± NaN|-8.52  | -9.25 ± 0.25| -0.63 | NaN ± NaN
C  |  methylthiol    | CS             |0|-1.20 ± 0.20| NaN ± NaN| NaN   | -0.99 ± 0.07| -1.42 | NaN ± NaN
"""

split_dat = re.split('\||\±|\n', data_str[1:-1])
split_dat = [word.strip() for word in split_dat]
```

```python
col_num = 14
all_dat_arr = np.array(split_dat)
all_dat_arr.shape = (int(len(all_dat_arr)/col_num), col_num)
```

```python
free_df = pd.DataFrame(all_dat_arr)
free_df.columns = ['res', 'name', 'smile', 'charge',
                   'wat_exp', 'wat_exp_sd', 'wat_sim', 'wat_sim_sd',
                   'oct_exp', 'oct_sim', 'oct_sim_sd',
                   'logp_exp', 'logp_sim', 'logp_sim_sd']
numeric_col = free_df.columns.values[3:]
free_df[free_df == 'NaN'] = np.nan
# Change column format to numeric
for col in numeric_col:
    free_df[col] = pd.to_numeric(free_df[col])
```

```python
free_df
```

## RMSE

```python
oct_RMSE = (((free_df['oct_exp']-free_df['oct_sim'])**2).mean())**0.5
oct_RMSE
```

```python
fig, ax = plt.subplots(1, figsize=(6,6))

#axs[0].errorbar(oct_exp, oct_sim, oct_sim_err,  fmt='o')

ax.errorbar(x=free_df.oct_exp, y=free_df.oct_sim, yerr=free_df.oct_sim_sd, fmt='.', c='black')

x = np.array([min(free_df.oct_exp)-1, max(free_df.oct_exp)+1])
gap = 1.0
ax.plot(x, x, color='black', linestyle='--')
ax.fill_between(x, x+gap, x-gap, color = 'k', alpha = 0.1)
ax.fill_between(x, x+2*gap, x-2*gap, color = 'blue', alpha = 0.05)
ax.grid()
ax.set_ylabel(r'$\Delta G^{sim} (kcal\;mol^{-1})$')
ax.set_xlabel(r'$\Delta G^{exp} (kcal\;mol^{-1})$')
ax.set_title(f'$RMSE = {oct_RMSE:.2f}\; kcal \; mol^{-1}$')
```
